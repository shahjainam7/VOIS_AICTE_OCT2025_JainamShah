**Cleaning Data**

```python
# Drop Duplicate Records
df.drop_duplicates(inplace=True)
```

```python
# Drop house rules and license with insufficient data
df.drop(['house_rules', 'license'], axis=1, inplace=True)
```

```python
# Remove all $ sign in price and service fee
df['price'] = df['price'].replace('$','', regex=False)
df['service fee'] = df['service fee'].replace('$','', regex=False)
```

```python
# Remove all comma sign in price and service fee
df['price'] = df['price'].replace(',','', regex=False)
df['service fee'] = df['service fee'].replace(',','', regex=False)
```

```python
# Rename price and service fee column to include a $ sign
df.rename(columns = {
    'price' : 'price_$',
    'service fee' : 'service_fee_$'
    },inplace=True)
```

```python
# Drop records with missing values
df.dropna(inplace=True)
```

```python
# Changing all mismatched datatypes to appropriate
df['price_$'] = df['price_$'].astype(float)
df['service_fee_$'] = df['service_fee_$'].astype(float)
df['id'] = df['id'].astype(str)
df['host id'] = df['host id'].astype(str)
df['last review'] = pd.to_datetime(df['last review'])
df['Construction year'] =  df['Construction year'].astype(int)
```

```python
# Correct the spelling of 'brookln' --> 'Brooklyn'
df.loc[df['neighbourhood group'] == 'brookln', 'neighbourhood group'] = 'Brooklyn'
```

```python
# Get rid of any outliers in 'availability 365' column data
df = df.drop(df[df['availability 365'] > 500].index)
```

**Exploratory Data Analysis (EDA)**

```python
df.describe()
```

```python
# Different types in dataset
property_types = df['room type'].value_counts().to_frame()
property_types
```

```python
# Room types plotted on barchart
room_type_bar = plt.bar(property_types.index, property_types.loc[:, "count"]);
plt.bar_label(room_type_bar, label = property_types.loc[:, "count"], padding = 4);
plt.title("Property type and Count");
plt.xlabel("Room Types");
plt.ylabel("Room Type Count");
```

```python
hood_group = df['neighbourhood group'].value_counts().to_frame()
hood_group
```

```python
# Neighbourhood group with highest number of listings
hood_group_bar = plt.bar(hood_group.index, hood_group.loc[:, "count"]);
plt.bar_label(hood_group_bar, label = hood_group.loc[:, "count"], padding = 4);
plt.ylim([0,40000]);
plt.xlabel("Neighbourhood Groups");
plt.ylabel("Number of Listings");
plt.xticks(rotation=45);
plt.title("Which neighbourhood has higher number of listings");
```

```python
# neighbourhood with highest average price for Airbnb listing
avg_price = df.groupby('neighbourhood group')['price_$'].mean().sort_values(ascending=False).to_frame()
avg_price_bar = plt.bar(avg_price.index, avg_price.loc[:, "price_$"]);
plt.bar_label(avg_price_bar, label = round(avg_price.loc[:, 'price_$'], 2), label_type = 'edge', padding = 4);
plt.ylim([0,700]);
plt.xlabel("Neighbourhood Groups");
plt.ylabel("Average Price per Listings ($)");
plt.xticks(rotation=45);
plt.title("Average Price per Listings ($) in each Neighbourhood Group");
```

```python
# Relation between Construction year of property and price
df.groupby('neighbourhood group')['price_$'].mean().to_frame().plot();
plt.xlabel('Construction Year');
plt.ylabel('Average Price ($)');
plt.title('Relation between Construction year of property and price');
```

```python
# Top 10 host by host listing count
hosts = df.groupby('host name')['calculated host listings count'].sum().sort_values(ascending = False).nlargest(10).to_frame()
hosts_bar = plt.bar(hosts.index, hosts.loc[:, 'calculated host listings count']);
plt.bar_label(hosts_bar, label = hosts.loc[:, 'calculated host listings count'], label_type='edge' ,padding = 3);
plt.xticks(rotation=80);
plt.ylim([0,120000]);
plt.xlabel('Host Name');
plt.ylabel('Host Listing Count');
plt.title('Top 10 Host by Host Listing Count');
```

```python
# Hosts verifies with identity more likely to give positive rating
review = df.groupby('host_identity_verified')['review rate number'].mean().sort_values(ascending=False).to_frame()
review
```

```python
review_bar = plt.bar(review.index, review.loc[:,"review rate number"]);
plt.bar_label(review_bar, label = round(review.loc[:,"review rate number"], 2), padding = 4);
plt.ylim([0,4]);
plt.xlabel('Host Identity Verified');
plt.ylabel('Average Review Rate');
plt.title('Review Rate for each verification status');
```

```python
base_color = sns.color_palette()[0]
sns.boxplot(data=df, x='host_identity_verified', y='review rate number', color = base_color);
plt.xlabel('Host verified status');
plt.ylabel('Review Rate');
plt.title('Review Rate for each verification status');
```

```python
# correlation between price of listing and its service fee
df['price_$'].corr(df['service_fee_$'])
```
